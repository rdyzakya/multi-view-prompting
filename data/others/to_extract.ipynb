```python
import pandas as pd

dataset = ["casa_absa-prosa", "hoasa_absa-airy"]
split = ["train", "valid", "test"]

# df = pd.read_csv(r"D:\Second Brain\Projects\Paper IJEEI\multi-view-prompting\data\others\casa_absa-prosa\train_preprocess.csv")

for ds in dataset:
    texts = []
    split_info = []
    for s in split:
        path = f"{ds}/{s}_preprocess.csv"
        df = pd.read_csv(path)
        texts.extend(df.iloc[:,0].values)
        split_info.extend([s for _ in range(df.shape[0])])
    texts = "####[([0], [1], 'POS')]\n".join(texts) + "####[([0], [1], 'POS')]"
    split_info = "\n".join(split_info)

    with open(f"{ds}_data.txt", 'w') as fp:
        fp.write(texts)
    
    with open(f"{ds}_split_info.txt", 'w') as fp:
        fp.write(split_info)
```
